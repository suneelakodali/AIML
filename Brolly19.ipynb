```python
import pandas as pd
```

```python
names = ['mpg', 'cylinders', 'diplacement', 'horsepower', 'weight', 'acceleartion', 'model_year', 'origin', 'car_name']
```

```python
df =pd.read_csv('auto-mpg.data',names=names,delim_whitespace=True)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df = df[df['horsepower'] != '?']
```

```python
df.shape
```

```python
df['horsepower'] = df['horsepower'].astype('float')
```

```python
df.dtypes
```

```python
df.columns
```

```python
X = df[['cylinders', 'diplacement', 'horsepower', 'weight','acceleartion', 'model_year', 'origin']]
y = df['mpg']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=32)
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
df1 = pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
df1
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```
